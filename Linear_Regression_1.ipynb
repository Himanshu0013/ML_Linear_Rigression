# Linear Regression_01

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.DataFrame({'No. of products' : [20,34 , 13, 30, 23],
                    'Raw_Material' : [20000, 35000, 10000, 30000, 21000],
                    'Sales' : [50000, 58000, 40000, 13000, 23000],
                    'revenue' : [100000, 150000, 200000, 500000, 800000]})
```

```python
data
```

```python
X = data[['No. of products','Raw_Material','Sales']]
X
```

```python
Y = data['revenue']
Y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,Y)
```

```python
pred_y_hat = lr.predict(X)
pred_y_hat
```

```python
from sklearn.metrics import r2_score
```

```python
accuracy = r2_score(Y,pred_y_hat)
accuracy
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(Y,pred_y_hat)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
mae = mean_absolute_error(Y,pred_y_hat)
mae
```
